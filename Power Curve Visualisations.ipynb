```python
from power_curve import PowerCurveAnalyzer
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Filter out tuples where the power value is 0 and format durations
def filter_powers(power_curve):

    filtered_durations = []
    filtered_powers = []
    tested_ftp = 241

    for duration, power in power_curve:
        if power > 0:
            filtered_durations.append(pca.format_duration(duration))
            filtered_powers.append(power)

    return filtered_durations, filtered_powers
```

```python
def plot_power_curve(power_curve, tested_ftp):

    filtered_durations, filtered_powers = filter_powers(power_curve)
    
    plt.figure(figsize=(10, 6))
    plt.plot(filtered_durations, filtered_powers, marker='o')
    plt.axhline(y=tested_ftp, color='g', linestyle='--', label='Tested FTP')
    plt.xlabel('Duration')
    plt.ylabel('Power (Watts)')
    plt.title('Power Curve')
    plt.show()
```

```python
# Visualise Power curve to Sep 2022 when tested FTP was 241W
pca = PowerCurveAnalyzer(data_source=None)
power_curve = pca.create_power_curve('Ride', '2022-09-30', 90)
plot_power_curve(power_curve, 241)
```

```python
# Visualise Power curve to Sep 2022 when tested FTP was 241W
pca = PowerCurveAnalyzer(data_source=None)
power_curve = pca.create_power_curve('Ride', '2022-09-30', 30)
plot_power_curve(power_curve, 241)
```
